```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Set plot style and size
plt.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (12,8)

# Reading data using pandas
df = pd.read_excel(r'C:\Users\NDLELA SEC SHOOL\Downloads\movies.xlsx')
```

```python
# looking at the data
df.head()
```

```python
# Missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Data types of columns
df.dtypes
```

```python
# Converting budget and gross columns to integers
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').astype('Int64')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').astype('Int64')
```

```python
df.head()
```

```python
# Extracting year from 'released' column
df['yearcorrected'] = df['released'].str.split(',').str[1].str[:5]
```

```python
df.head()
```

```python
# Sort DataFrame by 'gross' column in descending order
df.sort_values(by='gross', inplace=True, ascending=False)
```

```python
df.head()
```

```python
# All the data
pd.set_option('display.max_rows', None)
```

```python
# removing duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
# Replace 'NAType' values with NaN
df.replace('NAType', np.nan, inplace=True)

# Convert 'budget' and 'gross' columns to numeric data types
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')

# Scatter plot of budget vs gross earnings
plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Using seaborn for budget vs gross
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"black"}, line_kws={"color":"red"})
```

```python
# Looking at correlation(pearson, kendall, spearman)
#pearson method by default
df.corr()
```

```python
#Kendall method
df.corr(method='kendall')
```

```python
#Spearman method
df.corr(method='spearman')
```

```python
# Visualise the data
correlation_matrix=df.corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looking at Company
df.head()
```

```python
#Converting Categorical Data to unique Numeric in Dataframe
df_numeric=df
for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype=='object'):
        df_numeric[col_name]=df_numeric[col_name].astype('category')
        df_numeric[col_name]=df_numeric[col_name].cat.codes

df_numeric
```

```python
# Visualise the data in numeric form
correlation_matrix=df_numeric.corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numeric.corr()
```

```python
#Extracting Correlation Pairs from Correlation Matrix in Dataframe
correlation_mat=df_numeric.corr()
corr_pairs=correlation_mat.unstack()
corr_pairs
```

```python
#Sorting Correlation Pairs in Dataframe
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
#Filtering High Correlation Pairs in Dataframe
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
